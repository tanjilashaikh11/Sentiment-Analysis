```python
import pandas as pd

# Read the CSV file with the correct delimiter
df = pd.read_csv('sentiment-analysis.csv', delimiter=',', quotechar='"')

# Display the first few rows and columns
print(df.head())
print(df.columns)
```

```python
# Count of each sentiment category
sentiment_counts = df['Sentiment'].value_counts()
print(sentiment_counts)
```

```python
import matplotlib.pyplot as plt

# Visualizing the sentiment distribution
sentiment_counts.plot(kind='bar', color=['blue', 'red'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Convert 'Date/Time' to datetime format
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='%d/%m/%y %H:%M')

# Group by date and count sentiments
daily_sentiment = df.groupby(df['Date/Time'].dt.date)['Sentiment'].value_counts().unstack().fillna(0)

# Plotting sentiment trends over time
daily_sentiment.plot(kind='line', marker='o')
plt.title('Sentiment Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid()
plt.legend(title='Sentiment')
plt.show()
```

```python
# Box plot for confidence scores by sentiment
plt.figure(figsize=(10, 5))
df.boxplot(column='Confidence Score', by='Sentiment', grid=False)
plt.title('Confidence Score Distribution by Sentiment')
plt.suptitle('')
plt.xlabel('Sentiment')
plt.ylabel('Confidence Score')
plt.show()
```

```python
# Count of sentiments by source
source_sentiment_counts = df.groupby('Source')['Sentiment'].value_counts().unstack().fillna(0)
print(source_sentiment_counts)

# Plotting sentiment counts by source
source_sentiment_counts.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title('Sentiment Count by Source')
plt.xlabel('Source')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.show()
```

```python
import seaborn as sns

# Visualizing confidence score distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Confidence Score'], bins=30, kde=True)
plt.title('Confidence Score Distribution')
plt.xlabel('Confidence Score')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Map sentiments to numerical values for correlation analysis
df['Sentiment_Numeric'] = df['Sentiment'].map({'Positive': 1, 'Negative': 0})

# Correlation matrix
correlation = df[['Confidence Score', 'Sentiment_Numeric']].corr()
print(correlation)

# Heatmap for correlation
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Count of sentiments by location
location_sentiment_counts = df.groupby('Location')['Sentiment'].value_counts().unstack().fillna(0)

# Plotting sentiment counts by location
location_sentiment_counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Sentiment Count by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')
plt.show()
```

```python
from wordcloud import WordCloud

# Create a word cloud for positive sentiments
positive_words = ' '.join(df[df['Sentiment'] == 'Positive']['Text'])
wordcloud_positive = WordCloud(width=800, height=400, background_color='white').generate(positive_words)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_positive, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Positive Sentiments')
plt.show()

# Create a word cloud for negative sentiments
negative_words = ' '.join(df[df['Sentiment'] == 'Negative']['Text'])
wordcloud_negative = WordCloud(width=800, height=400, background_color='black', colormap='plasma').generate(negative_words)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_negative, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Negative Sentiments')
plt.show()
```

```python
# Resampling data to get daily counts
daily_counts = df.set_index('Date/Time').resample('D').count()['Sentiment']

# Plotting daily counts
plt.figure(figsize=(12, 6))
daily_counts.plot()
plt.title('Daily Count of Sentiments')
plt.xlabel('Date')
plt.ylabel('Count')
plt.grid()
plt.show()
```
